```python
import os
import re
import fitz  # PyMuPDF
import pandas as pd

# Function to extract text from a PDF
def pdf_to_text(pdf_path):
    try:
        pdf_document = fitz.open(pdf_path)
        extracted_text = ""
        for page_num in range(len(pdf_document)):
            page = pdf_document.load_page(page_num)
            text = page.get_text()
            extracted_text += text
        return extracted_text
    except Exception as e:
        print(f"Error reading PDF file: {e}")
        return ""

# Function to clean and extract information from text
def extract_invoice_info(text):
    info = {}

    # Split text into lines
    lines = text.split('\n')

    # Regular expressions for different fields
    invoice_regex = re.compile(r"Invoice[:\s#]*(\w+)")
    bill_to_regex = re.compile(r"Bill To[:\s]*(.+)")
    ship_to_regex = re.compile(r"Ship To[:\s]*(.+)")
    date_regex = re.compile(r"(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s+\d{2}\s+\d{4}")
    balance_due_regex = re.compile(r"Balance Due[:\s]*([0-9,]+\.\d{2})")
    total_regex = re.compile(r"Total[:\s]*([0-9,]+\.\d{2})")
    order_id_regex = re.compile(r"Order ID[:\s#]*(\w+)")

    # Initialize variables to store extracted data
    invoice_number = None
    bill_to = None
    ship_to = None
    date = None
    balance_due = None
    total = None
    order_id = None

    items = []

    # Iterate through each line to match patterns and extract data
    for i, line in enumerate(lines):
        # Extract invoice number from line below header
        if "INVOICE" in line and i + 1 < len(lines):
            match = invoice_regex.search(lines[i + 1])
            if match:
                invoice_number = match.group(1).strip()

        # Extract Bill To and Ship To from line below header
        if "Bill To" in line and i + 1 < len(lines):
            bill_to = lines[i + 1].strip()

        if "Ship To" in line and i + 1 < len(lines):
            ship_to = lines[i + 1].strip()

        # Extract Date and Balance Due from the same line
        date_match = date_regex.search(line)
        if date_match:
            date = date_match.group(0).strip()

        balance_due_match = balance_due_regex.search(line)
        if balance_due_match:
            balance_due = balance_due_match.group(1).strip()

        # Extract Total from the same line
        total_match = total_regex.search(line)
        if total_match:
            total = total_match.group(1).strip()

        # Extract Order ID from the same line
        order_id_match = order_id_regex.search(line)
        if order_id_match:
            order_id = order_id_match.group(1).strip()

        # Extract item details
        if "Item" in line and "Quantity" in line and "Rate" in line and "Amount" in line:
            for item_line in lines[i + 1:]:
                if "Total" in item_line or "Balance Due" in item_line or not item_line.strip():
                    break  # Stop reading if "Total" or an empty line is encountered

                item_parts = re.split(r'\s{2,}', item_line.strip())
                if len(item_parts) >= 4:
                    item_info = {
                        "Item": item_parts[0].strip(),
                        "Quantity": item_parts[1].strip(),
                        "Rate": item_parts[2].strip(),
                        "Amount": item_parts[3].strip()
                    }
                    items.append(item_info)

    # Store extracted information in the dictionary
    info['Invoice Number'] = invoice_number or "Not Found"
    info['Bill To'] = bill_to or "Not Found"
    info['Ship To'] = ship_to or "Not Found"
    info['Date'] = date or "Not Found"
    info['Balance Due'] = balance_due or "Not Found"
    info['Total'] = total or "Not Found"
    info['Order ID'] = order_id or "Not Found"
    info['Items'] = items if items else [{"Item": "Not Found", "Quantity": "Not Found", "Rate": "Not Found", "Amount": "Not Found"}]

    return info

# Specify the path to the PDF file
pdf_path = 'invoice_Aaron Bergman_36258.pdf'

# Extract text from the PDF
text = pdf_to_text(pdf_path)

# Check if text extraction was successful
if text:
    # Extract invoice information
    invoice_info = extract_invoice_info(text)

    # Prepare the data for CSV
    data = {
        "Invoice Number": [invoice_info['Invoice Number']],
        "Bill To": [invoice_info['Bill To']],
        "Ship To": [invoice_info['Ship To']],
        "Date": [invoice_info['Date']],
        "Balance Due": [invoice_info['Balance Due']],
        "Total": [invoice_info['Total']],
        "Order ID": [invoice_info['Order ID']]
    }

    # Create a DataFrame for the main information
    df_main = pd.DataFrame(data)

    # Create a DataFrame for the items
    df_items = pd.DataFrame(invoice_info['Items'])

    # Save the main information to a CSV file
    output_path_main = 'output.csv'
    df_main.to_csv(output_path_main, index=False)

    # Save the items to a CSV file
    output_path_items = 'output.csv'
    df_items.to_csv(output_path_items, index=False)

    print(f"Main invoice information saved to {output_path_main}")
    print(f"Item details saved to {output_path_items}")
else:
    print("Failed to extract text from the PDF.")
```
